## Pipeline: Clean Continuous Features

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

This dataset contains information about 891 people who were on board the ship when departed on April 15th, 1912. As noted in the description on Kaggle's website, some people aboard the ship were more likely to survive the wreck than others. There were not enough lifeboats for everybody so women, children, and the upper-class were prioritized. Using the information about these 891 passengers, the challenge is to build a model to predict which people would survive based on the following fields:

- **Name** (str) - Name of the passenger
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

**This notebook will implement some of the cleaning that was done in Section 2: EDA & Data Cleaning**

### Read in Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
titanic_df = pd.read_csv('../Data/titanic.csv')
titanic_df.head()
```

### Clean continuous variables
1. Fill in missing values for `Age`
2. Combine `SibSp` & `Parch`
3. Drop irrelevant/repetitive variables (`SibSp`, `Parch`, `PassengerId`)

#### Fill missing for `Age`

```python
# to see the total missing values for each columns
titanic_df.isnull().sum()
```

```python
# replace mean age value to the missing Ages
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
```

#### Combine `SibSp` & `Parch`

```python
# to see relationship of SibSp and Parch with Survival Rate
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic_df, kind='point', aspect=2)
```

```python
# as we already knew SibSp and Parch have similar relationship, we will combine those columns
titanic_df['Family_cnt'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df.head()
```

#### Drop unnnecessary variables

```python
titanic_df.drop(['PassengerId', 'SibSp', 'Parch'], axis=1, inplace=True)
```

### Write out cleaned data

```python
titanic_df.to_csv('../Data/titanic_cleaned.csv', index=False) # index false to make sure next time when pandas read this csv, to avoid assuming it as feature
```
