# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

----------

# Functional Methods

```python
plt.plot(x, y, color="red", linestyle="--");
plt.title("Sample Chart")
plt.xlabel("xlabel")
plt.ylabel("ylabel");
```

### Multi plots

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, "r");

plt.subplot(1, 2, 2)
plt.plot(y, x, "b");
```

---------

# OOP Method - better way

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
# left, bottom, width, height (range 0 to 1)
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

# Plot on that set of axes
axes.plot(x,y);
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Sample Chart");
```

```python
fig = plt.figure()

#left, bottom, width, height
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]);
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]);
axes2 = fig.add_axes([0.5, 0.2, 0.3, 0.2]);
```

```python
# Create Figure (empty canvas)
fig = plt.figure() 

# Add set of axes to figure
# left, bottom, width, height (range 0 to 1)
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]); # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]); # inset axes


# Larger Figure Axes 1
axes1.plot(x,y);
axes1.set_title("Larger One");
axes1.set_xlabel("Xlabel axes1")
axes1.set_ylabel("Ylabel axes1");

# Insert Figure Axes 2
axes2.plot(y,x);
axes2.set_title("Smaller One");
axes2.set_xlabel("Xlabel axes2")
axes2.set_ylabel("Ylabel axes2");
```

-------------

### Creating Subplots with OOP Methods

```python
fig, axes = plt.subplots(nrows=2, ncols = 3)

# take care of overlapping layout
plt.tight_layout();
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2)

for current_axes in axes:
    current_axes.plot(x,y);
```

```python
fig, axes = plt.subplots(nrows=1, ncols = 2)

axes[0].plot(x,y);
axes[0].set_title("first plot")

axes[1].plot(y,x);
axes[1].set_title("second plot");
```

----------

# Fig Size and DPI

```python
fig = plt.figure(figsize = (8,2)) #8 inches, 2 inches
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x,y);
```

```python
fig, ax = plt.subplots(figsize=(8,2))
ax.plot(x,y);
```

```python
fig, ax= plt.subplots(nrows=2, ncols=1, figsize=(8,2))

ax[0].plot(x,y);
ax[1].plot(y,x);

plt.tight_layout();
```

## Saving Figures

```python
fig.savefig("samplefig.png", dpi = 150)
```

## Legends

```python
fig, ax = plt.subplots()

ax.plot(x, x**2, label = "X squared")
ax.plot(x, x**3, label = "X cube");

ax.legend(); #legend looks for label for each plot
```

# Setting colors, linewidths, linetypes

## Plot Appearance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = "#FF8C00", linewidth = 5, alpha = 0.7, linestyle="--"); #RGB hex code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# can also use short form as keyword
ax.plot(x,y, color = "#FF8C00", lw = 5, alpha = 0.7, ls="steps"); 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y, color = "purple", lw = "2", ls = "-", marker = "o", markersize = "10", markerfacecolor = "yellow", markeredgewidth = "2", markeredgecolor = "green");
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1,])
ax.plot(x,y, color = "purple", lw = "2", ls="--");

#zooming in
ax.set_xlim([0,1]) #left bound, right bound
ax.set_ylim([0,2]) #upper bound, lower bound
```
