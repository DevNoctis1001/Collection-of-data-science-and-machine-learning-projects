# Time Series Generator

This class takes in a sequence of data-points gathered at
equal intervals, along with time series parameters such as
stride, length of history, etc., to produce batches for
training/validation.

#### Arguments
    data: Indexable generator (such as list or Numpy array)
        containing consecutive data points (timesteps).
        The data should be at 2D, and axis 0 is expected
        to be the time dimension.
    targets: Targets corresponding to timesteps in `data`.
        It should have same length as `data`.
    length: Length of the output sequences (in number of timesteps).
    sampling_rate: Period between successive individual timesteps
        within sequences. For rate `r`, timesteps
        `data[i]`, `data[i-r]`, ... `data[i - length]`
        are used for create a sample sequence.
    stride: Period between successive output sequences.
        For stride `s`, consecutive output samples would
        be centered around `data[i]`, `data[i+s]`, `data[i+2*s]`, etc.
    start_index: Data points earlier than `start_index` will not be used
        in the output sequences. This is useful to reserve part of the
        data for test or validation.
    end_index: Data points later than `end_index` will not be used
        in the output sequences. This is useful to reserve part of the
        data for test or validation.
    shuffle: Whether to shuffle output samples,
        or instead draw them in chronological order.
    reverse: Boolean: if `true`, timesteps in each output sample will be
        in reverse chronological order.
    batch_size: Number of timeseries samples in each batch
        (except maybe the last one).

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# define generator
length = 2 # Length of the output sequences (in number of timesteps)
batch_size = 1 #Number of timeseries samples in each batch

# in our case, data and targets are given same value. Because we are predicting time series and both i
generator = TimeseriesGenerator(data=scaled_train, targets=scaled_train,
                                                   length=length, batch_size=batch_size)
```

# Create a model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
# we need to define number of features, in our case there is only one feature; the value we are trying to predict
n_features = 1
```

```python
# define model
model = Sequential()

# Simple RNN layer
model.add(SimpleRNN(units=50, input_shape=(length, n_features)))

# Final Prediction
model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
# fit model using generator
model.fit(generator, epochs=5, verbose=1)
```

### Visualize the losses

# Remember!!!  for RNN, the prediction of one neuron is passed as input to next neuron together with feature input.

# these range of inputs are like window slider as in SQL (sliding one window to another window, only one point at a time.) 
+ So the very first data point from previous window will be removed.
+ and one new point will be added this this current window.
+ Note: stepping one point, etc is based on how we configured our model. This can be changed. In this example, we are moving 1 step at at time.
    + predic one value based on a sequence of inputs
    + create a new sequence of input data points
        + drop the very first data point
        + add the newly predicted data point into the end of the sequences
        + use this newly created sequence to predict a next data point.
        + loop the process.

```python
test_predictions = []

first_evaluation_batch = scaled_train[-length: ]

# saying we want to predict 1 sample data point, using a input sequences of 50 data points, with 1 feature.
current_batch = first_evaluation_batch.reshape(1, length, n_features) 
```

```python
test_predictions = [] # this is to keep all the predictions

first_evaluation_batch = scaled_train[-length: ]

# saying we want to predict 1 sample data point, using a input sequences of 50 data points, with 1 feature.
current_batch = first_evaluation_batch.reshape((1, length, n_features))

for i in range(len(test)):
    current_prediction = model.predict(current_batch)[0] # make the prediction and get the predicted value
   
    test_predictions.append(current_prediction)
    
    current_batch = np.append(current_batch[:, 1:, :], [[current_prediction]], axis=1) # drop first data point, append newly predicted value at the end, reasign back
```

# Forecasting

Forecast into unknown range. We should first utilize all our data, since we are now forecasting!

```python
# scale the full data set
full_scaler = MinMaxScaler()
scaled_full_data = full_scaler.fit_transform(df)
```

```python
# generator
length = 50 # Length of the output sequences (in number of timesteps)

generator = TimeseriesGenerator(scaled_full_data, scaled_full_data, length=length, batch_size=1)
```

```python
# create model
model = Sequential()

model.add(LSTM(50, input_shape=(length, n_features)))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
# fit the model
model.fit(generator, 
              epochs=6,
              verbose=1)
```

```python
# make forecast into the future
forecast = []

first_evaluation_batch = scaled_full_data[-length: ]
current_batch = first_evaluation_batch.reshape((1, length, n_features))

for i in range(len(test)):
    current_prediction = model.predict(current_batch)[0]
    
    forecast.append(current_prediction)
    
    current_batch = np.append(current_batch[:, 1: , :], [[current_prediction]], axis=1)
```

```python
# inverse transform the forecast values
true_forecast = full_scaler.inverse_transform(forecast)
```
