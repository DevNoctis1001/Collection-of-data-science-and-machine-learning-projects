## Foundations: Clean Data

Using the Titanic dataset from [this](https://www.kaggle.com/c/titanic/overview) Kaggle competition.

This dataset contains information about 891 people who were on board the ship when departed on April 15th, 1912. As noted in the description on Kaggle's website, some people aboard the ship were more likely to survive the wreck than others. There were not enough lifeboats for everybody so women, children, and the upper-class were prioritized. Using the information about these 891 passengers, the challenge is to build a model to predict which people would survive based on the following fields:

- **Name** (str) - Name of the passenger
- **Pclass** (int) - Ticket class
- **Sex** (str) - Sex of the passenger
- **Age** (float) - Age in years
- **SibSp** (int) - Number of siblings and spouses aboard
- **Parch** (int) - Number of parents and children aboard
- **Ticket** (str) - Ticket number
- **Fare** (float) - Passenger fare
- **Cabin** (str) - Cabin number
- **Embarked** (str) - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

### Read in Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
titanic_df = pd.read_csv('../Data/titanic.csv')
titanic_df.head()
```

### Clean continuous variables

#### Fill missing for `Age`

```python
titanic_df.isnull().sum()
```

```python
# fill missing Age with mean value
titanic_df['Age'].fillna(titanic_df['Age'].mean(), inplace=True)
titanic_df.head(10)
```

#### Combine `SibSp` & `Parch`

```python
for i, col in enumerate(['SibSp', 'Parch']):
    plt.figure(i)
    sns.catplot(x=col, y='Survived', data=titanic_df, kind='point', aspect=2, )
```

#### since those above features are closely related and similare trajectory, we will combine them as one feature

```python
titanic_df['Family_cnt'] = titanic_df['SibSp'] + titanic_df['Parch']
```

#### Drop unnnecessary variables

```python
titanic_df.drop(['SibSp', 'Parch', 'PassengerId'], axis=1, inplace=True)
```

```python
titanic_df.head(10)
```

### Clean categorical variables

#### Fill in missing & create indicator for `Cabin`

```python
titanic_df.isnull().sum()
```

```python
titanic_df.groupby(titanic_df['Cabin'].isnull())['Survived'].mean()
```

As we see in above that survival rate of people with cabin missing is around 30% and pepole with cabin is around 67%. This led us to belive that cabin was not missing at random. In fact, those people actually didn't have a cabin. As a result, they are very unlikely to survive.

so we want to make sure to capture this information and encode thie variable.

```python
titanic_df['Cabin_ind'] = np.where(titanic_df['Cabin'].isnull(), 0, 1) #condition, if , else
titanic_df.head()
```

#### Convert `Sex` to numeric

```python
gender_num = {
    'male': 0,
    'female': 1
}

titanic_df['Sex']  = titanic_df['Sex'].map(gender_num)
titanic_df.head()
```

#### Drop unnecessary variables

```python
titanic_df.drop(['Cabin', 'Name', 'Ticket', 'Embarked'], axis=1, inplace=True)

titanic_df.head()
```

### Write out cleaned data

```python
titanic_df.to_csv('../Data/titanic_cleaned.csv', index=False)
```
