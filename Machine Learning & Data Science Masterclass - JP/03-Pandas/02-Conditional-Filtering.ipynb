# Conditional Filtering

```python
import numpy as np
import pandas as pd

tips = pd.read_csv('../Data/tips.csv')
tips.head()
```

## One Condition

```python
# total bill more than 40 $
boolean_series = tips['total_bill'] > 40
boolean_series
```

```python
tips[boolean_series]
```

```python
# same as above 2 lines
tips[tips['total_bill'] > 40]
```

```python
# male cusomters only
tips[tips['sex'] == 'Male']
```

## Multiple Conditions
- OR `|` (either condition is True)
- AND `&` (both conditions must be True)
- `~` (not operator)

```python
tips[(tips['total_bill'] > 40) & (tips['sex'] == 'Male')]
```

```python
tips[(tips['total_bill'] > 40) | (tips['sex'] == 'Male')]
```

```python
tips['day'].value_counts()
```

```python
tips[(tips['day'] == 'Sat') | (tips['day'] == 'Sun')]
```

```python
tips[tips['day'] != 'Sun']
```

```python
tips[~(tips['day']=='Sun')]
```

### using `isin` for multiple conditions
- OR multiple conditions

```python
options = ['Sat', 'Sun']
```

```python
tips['day'].isin(options)
```

```python
tips[tips['day'].isin(options)]
```
