# Grid Search
- is capable of testing a dictionary of multiple hyperparameters options through cross-validation.
- allows for both Cross Validation and Grid Search to be performed in a generalized way for any model.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Data/Advertising.csv')
```

```python
df.head()
```

```python
# Create X and Y
X = df.drop('sales', axis=1)
y = df['sales']

# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# Scale Data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Model

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_net_model = ElasticNet()
```

## Grid Search

A search consists of:

* an estimator (regressor or classifier such as sklearn.svm.SVC());
* a parameter space;
* a method for searching or sampling candidates;
* a cross-validation scheme 
* a score function.

```python
parameters = {
    'alpha': [0.1, 1, 5, 10, 50, 100],
    'l1_ratio': [.1, .5, .7, .95, .99, 1],
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_net_model,
                                             param_grid=parameters,
                                             scoring='neg_mean_squared_error',
                                             cv=5, verbose=2)
```

```python
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

# Using Best Model From Grid Search

For GridSearchCV, the result is already trained model with best parameter

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
