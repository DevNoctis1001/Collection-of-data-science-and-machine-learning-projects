# Matplotlib Figure Object

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0, 10, 11)
```

```python
b = a ** 4
```

```python
a, b
```

```python
x = np.arange(0, 10)
y = x * 2
```

```python
x, y
```

```python
fig = plt.figure()

#Large axes
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('power of 4')


#Small axes
axes2 = fig.add_axes([0.2, 0.35, 0.5, 0.5])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoom in')

plt.show()
```

# Figure Parameters

# adding figsize, dpi, binding box 
- bbox_inches => to make sure the boundaries are also saved in the figure

```python
fig = plt.figure(figsize=(2,2), dpi= 100)

axes1 = fig.add_axes([0, 0, 1, 1])

axes1.plot(a, b)

fig.savefig('Figures/new_figure.png', bbox_inches='tight') 
```
