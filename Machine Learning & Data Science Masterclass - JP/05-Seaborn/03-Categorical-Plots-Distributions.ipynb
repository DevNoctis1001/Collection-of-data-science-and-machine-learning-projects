# Categorical Plots - Distribution within Categories

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/StudentsPerformance.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

-----

# Boxplot

```python
plt.figure(figsize=(12,5))
plt.title('boxplot')

sns.boxplot(data=df, y='math score', x='test preparation course');
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df, y='reading score', x='parental level of education');
```

# Adding hue for further segmentation
- hue add more subcategory into the categeory

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df, y='reading score', x='parental level of education', hue='test preparation course');

plt.legend(bbox_to_anchor=(1.05, 1)); # to move the legend
```

```python
plt.figure(figsize=(12,5))
sns.boxplot(data=df, x='reading score', y='parental level of education', hue='test preparation course', palette='Set2');

plt.legend(bbox_to_anchor=(1.05, 1)); # to move the legend
```

# Boxplot Styling Parameters


#### Orientation

```python
sns.boxplot(data=df, x='math score', y='parental level of education', orient='h')
```

#### Width

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, y='math score', x='parental level of education', hue='gender', width=0.5)
```

-------

# ViolinPlot

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='math score', y='parental level of education');
```

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='parental level of education',  y='math score', hue='gender');

plt.legend(bbox_to_anchor=(1.2, 1))
```

# Violinplot Parameters

# Split
- it split the the distribution into two parts based on the hue. You cans see different colors and compare them
- However it maybe much easier to use boxplot rather than violinplot.

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='parental level of education',  y='math score', hue='gender', split=True);

plt.legend(bbox_to_anchor=(1.2, 1))
```

# Inner

## Inner = None
- get rid of inner black box and white tiny median point in the middle

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, x='parental level of education',  y='math score', inner=None);
```

# Inner = 'quartile'

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, y='parental level of education',  x='math score', inner='quartile');
```

# Inner='stick'
- draw a stick for every instance

```python
plt.figure(figsize=(12,8))
sns.violinplot(data=df, y='parental level of education',  x='math score', inner='stick');
```

# bandwidth (bw)
- smaller value means more sensitivity like in KDE

```python
plt.figure(figsize=(12,4))
sns.violinplot(data=df, y='parental level of education',  x='math score', bw=0.1);
```

--------

# Swarmplot

seaborn will let us know everypoint can't be fit into the plot.

```python
sns.swarmplot(data=df, x='math score')
```

to fit every point, we can make point to get smaller. but depends on the use case.
# size

```python
plt.figure(figsize=(12,5))
sns.swarmplot(data=df, x='math score', size=2)
```

```python
plt.figure(figsize=(12,5))
sns.swarmplot(data=df, x='math score', y='gender', size=2)
```

# dodge=True for Hue

```python
plt.figure(figsize=(12,5))
sns.swarmplot(data=df, x='math score', y='gender', size=2, hue='test preparation course', dodge=True)

plt.legend(bbox_to_anchor=(1.15, 1))
```
