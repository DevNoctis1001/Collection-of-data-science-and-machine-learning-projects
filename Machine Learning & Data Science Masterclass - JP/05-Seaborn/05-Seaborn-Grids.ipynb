```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/StudentsPerformance.csv')
df.head()
```

# Catplot() - categorical plot
Catplot is `very base level call` and allows us to use `any categorical plot`. 
Catplot allows us to use `FacetGrid` on top of which kind of plot.

**NOTE: use catplot() only if you want to use row, col. Otherwise, use highlevel abstractions such as sns.violinplot() directly for easier understanding.**

Kind Options are: 
- “point"
- “bar”, 
- “strip”
- “swarm”
- “box”
- “violin”
- “boxen”

```python
# in this case, we are using catplot to draw boxplot
sns.catplot(data=df, x='gender', y='math score', kind='box')
```

## row='feature-name'
- row is like using `hue`. but display on the different row.

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', row='lunch')
```

As we can see `row` parameter does the similar thing as `hue`, but displays on different row as it supposed to.

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', hue='lunch')
```

## col = 'feature-name'
- similar to row and display on different columns.

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch')
```

## row + col

```python
sns.catplot(data=df, x='gender', y='math score', kind='box', col='lunch', row='test preparation course')
```

```python
sns.catplot(data=df, x='gender', y='math score', kind='violin')
```

# PairGrid
- parigrid is the behind the scence of `pairplot`, which allows us to use `grid system`

```python
sns.PairGrid(df)
```

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot) # map scatterplot on the upper diagional
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.histplot)

g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df)

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)
```

```python
g = sns.PairGrid(df, hue='gender')

g = g.map_upper(sns.scatterplot)

g = g.map_diag(sns.kdeplot)

g = g.map_lower(sns.kdeplot)

g = g.add_legend()
```
