# Example

```python
np.random.seed(42)
sample_ages = np.random.randint(0, 100, 200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages.head(2) 
```

```python
sns.set(style='white')

sns.rugplot(x='age', data=sample_ages, height=0.5)
```

```python
sns.displot(data=sample_ages, x='age', rug=True, bins=30, kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age')
```

# Cut Off KDE
#### clip for clipping which point from which point

```python
#if we want the cut off from 0 to 100 like in displot
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100])
```

# Bandwidth
the KDE is constructed through the summation of the kernel (most commonly Gaussian), we can effect the bandwith of this kernel to make the KDE more "sensitive" to the data. Notice how with a smaller bandwith, the kernels don't stretch so wide, meaning we don't need the cut-off anymore. This is analagous to increasing the number of bins in a histogram (making the actual bins narrower).
- the lower the bw_adjust value, the higher the bins (more sensitive)

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.01)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.1)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.5)
```

```python
sns.kdeplot(data=sample_ages,x='age',bw_adjust=1)
```

# Basic Styling
- shade

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0, 100], bw_adjust=0.6, shade=True, color='green')
```
