# Categorical Plots - Statistical Estimation within Categories

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/dm_office_sales.csv')
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
df['level of education'].value_counts()
```

# Countplot()
- we can visualize the pandas `value_counts` info with countplot.

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='division')
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='level of education')
```

# Breakdown within another category with 'hue'

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='level of education', hue='division')
```

### changing palette
- https://matplotlib.org/3.3.3/tutorials/colors/colormaps.html#miscellaneous

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='level of education', hue='division', palette='Set2')
```

# barplot()
- similar to a `.groupby(x_axis).count()` call in pandas
- `estimator`: we need to use numpy avaliable aggregration functions.
- `ci`is confidence interval. (https://stackoverflow.com/questions/58362473/what-does-black-lines-on-a-seaborn-barplot-mean)

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd')
```

we can see in the below chart, master degree doesn't have black vertical bar. Because there are only 3 data and each of these are charted. so there is no way to calculate standard deviation.

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')
```

# moving legend

```python
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x='level of education', y='salary', estimator=np.mean, ci='sd', hue='division')

plt.legend(bbox_to_anchor=(1.05, 1))
```
