# Seaborn

Seaborn is using matplotlib under the hood. So we can use matplotlib for any customizations.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/dm_office_sales.csv')
```

```python
df.head()
```

```python
df.info()
```

# Scatter Plots

```python
# as seaborn is using matplotlib under the hood, we can use plt for customization
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df);
```

# Seaborn Parameters

# Hue

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education');
```

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, hue='salary');
```

# Choosing a palette from Matplotlib's cmap
- https://matplotlib.org/tutorials/colors/colormaps.html

# palette

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education', palette='viridis');
```

# size

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, size='work experience')
```

# if we want to make point size to be uniform and make it bigger
- use parameter s

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, s=200)
```

# alpha
- transparaency

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(x='salary', y='sales', data=df, size='work experience', alpha=0.5)
```

------

# Style - change the marker style

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, style='level of education', hue='level of education', s=150)

plt.savefig('Figures/level-of-eduation.png')
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, style='level of education', hue='level of education', s=150, alpha=0.4)
```
