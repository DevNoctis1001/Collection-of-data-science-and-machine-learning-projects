# Matrix Plots
- matrix plots are visual equivalent of `pivot table`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/country_table.csv')
df.head()
```

# Heatmap
- heatmap needs categorical index and feature column in numerical values
- it is important to have same unit for numeric values. for instance, values are in rate %.

```python
# set categorical index
df = df.set_index('Countries')
```

```python
df.head()
```

## importance of having same unit for numeric values
in this case, `Life expectancy` is unit of years and other columns are in the rate. Scales are completely different and units are different.

```python
sns.heatmap(data=df)
```

### so always make sure to drop those columns before hand

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1))
```

### linewidth & annot
- linewidth: to space out between border
- annot: annotation for values

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True)
```

### cmap for color mapping

```python
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis')
```

### center
- if we want to set a value to center for lower and upper boundaries
- in general, we don't really change it.

```python
# now 40 became center value and lighter or darker colors are upper and lower on that value
sns.heatmap(data=df.drop('Life expectancy', axis=1), linewidth=0.5, annot=True, cmap='viridis', center=40)
```

# Clustermap
- we can see the cluster with the line group

```python
sns.clustermap(df, annot=True)
```

### if clustering columns doesn't make sense for some cases, we can switch it off
- `col_cluster = False` 

```python
sns.clustermap(df, annot=True, col_cluster=False)
```

### cluster bar position
`cbar_pos`

we can see the legend bar is shifted.

```python
sns.clustermap(df, annot=True, col_cluster=False, cbar_pos=(-0.1, .2, .03, .4))
```
