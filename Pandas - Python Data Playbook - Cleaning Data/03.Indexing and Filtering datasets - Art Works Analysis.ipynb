# 3) Using `.iloc` to access Specific Rows or Columns
- Data Frame method, look at `integer position` of Rows and Columns

```python
data = pd.read_csv("Data/artwork_sample.csv")
data.head(2)
```

```python
data.iloc[0, :]
```

### Aware of the difference between `iloc` and `loc`

```python
# take note that iloc look at integer position.
# so it only return data from 0 to 2 (start, end - exclusive)
# it acts same as normal python start, stop position way
data.iloc[0:3, :]
```

```python
# if we use loc, this return row label number 0 to 3 (inclusive)
# because loc is looking at row label 
data.loc[0:3, :]
```

### replace row index with specific column

```python
data.set_index("id", inplace = True)
```

```python
data.head(2)
```

```python
# now we have changed row index label, we can clearly see the difference between loc and iloc
data.loc[1036:1038, :]
```

```python
data.iloc[1:4, :]
```

```python
# if we use the same integer postion for loc, it returns nothing
# because loc is looking atindex label and there is nothing with index label 1 to 4
data.loc[1:4, :]
```

----------

now using for columns index

```python
# as "id" is identified as index, it won't be counted in indexing
# that's why "accession_number" is 0, artist = 1, artistRole = 2
data.iloc[0:3, 0:3]
```

```python
# if we reset the index label and retry the code above again, we will have different result
# because now "id" column is no longer  row index and it became part of column again.
data.reset_index(inplace = True)
data.iloc[0:3, 0:3]
```

# 4) Filtering data with `str.contains()`

```python
data = pd.read_csv("Data/artwork_sample.csv")
data.head(2)
```

```python
data.medium
```

```python
type(data.medium)
```

```python
# for series, we can use .str method
data.medium.str
```

```python
data.medium.str.contains("Graphite")
```

```python
data[data.medium.str.contains("Graphite")]
```

```python
data.loc[data.medium.str.contains("Graphite"), ["artist", "medium"]]
```

```python
# when using case insensitvie
data.loc[data.medium.str.contains("Graphite", case=False), ["artist", "medium"]]
```

```python
data.loc[data.medium.str.contains("(?i)Graphite", regex = True), ["artist", "medium"]]
```

### multiple conditions

```python
data.loc[data.medium.str.contains("(?i)Graphite", regex = True) | data.medium.str.contains("(?i)Line", regex = True), ["artist", "medium"]]
```

```python
data.loc[data.medium.str.contains("Graphite|Line", case = False, regex = True), ["artist", "medium"]]
```
