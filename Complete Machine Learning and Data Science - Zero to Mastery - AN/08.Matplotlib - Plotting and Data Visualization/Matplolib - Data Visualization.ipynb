## 3 methods of plotting

```python
#1st method
fig = plt.figure() # creates a figure
ax = fig.add_subplot() # add some axes
plt.show()
```

```python
# 2nd method
fig = plt.figure() # creates a figure
ax = fig.add_axes([1,1,1,1])
ax.plot(x, y)
plt.show()
```

```python
# 3rd method (recommended one) / Object Oriented API
fig, ax = plt.subplots() #create figure and subplots
ax.plot(x, y);
```

```python
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
```

# Matplotlib example workflow

```python
# 0. import matplolib and get it ready for plotting in Jupyter
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

```python
# 2. Set up Plot
fig, ax = plt.subplots(figsize = (5, 5)) #(width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title = "Simple Plot",
           xlabel = "x-axis",
           ylabel = "y-axis")

# 5. Save & Show with others (you save the whole figure)
fig.savefig("images/sample-plot.png")
```

# Making figures with NumPy arrays
- Line Plot
- Scatter Plot
- Bar Plot
- Histogram
- Subplot

```python
import numpy as np
```

```python
# create data
x = np.linspace(0, 10, 100) #create 100 of evenly distributed data between 0 and 10
x
```

```python
# Set up & Plot the data & create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# create some data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10,
                     "Peanut Butter": 8,
                     "Cashew Butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Nut Butter Store",
          ylabel = "Price");
```

```python
#horizontal bar graph
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Nut Butter Store",
          xlabel = "Price");
```

```python
#Histogram
# make some data
x = np.random.randn(1000) #normal distributed random data

fig, ax = plt.subplots()
ax.hist(x);
```

# Two options for subplots


```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,
                                                                ncols = 2,
                                                                figsize=(10,5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                                 ncols = 2,
                                  figsize = (10, 5))

#plot to each different indexes
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```
