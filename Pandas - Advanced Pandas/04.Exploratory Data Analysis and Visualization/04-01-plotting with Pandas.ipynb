## <center> Plotting with Pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
daterange = pd.period_range(start='1/1/1950', freq='1d', periods=50)

date_df = pd.DataFrame(data=daterange, columns=['day'])
date_df['value1'] = np.random.randint(45, 64, size=len(date_df))
date_df['value2'] = np.random.randint(25,35,size=(len(date_df)))

date_df.head()
```

```python
ax = date_df.plot();
```

-----


# Area plot and Stacking

```python
date_df.plot.area(stacked=False); # note that y-axis value (because it is not stacked, blue shade values are in the range of 60s)
```

```python
date_df.plot.area(stacked=True); # note that y-axis value (because it is now stacked one upon another, value1 + value 2)
```

# Histogram

```python
iris = pd.read_csv('../data/iris.csv')
```

```python
iris.head()
```

```python
iris.hist();
```

-------

# Overlay Variables

```python
iris[['sepal_width', 'sepal_length']].plot.hist(alpha=0.5);
```

------

# Scatterplot

```python
colors = {
    'versicolor': 'red',
    'setosa': 'green',
    'virginica': 'blue'
}

iris['colors'] = iris['species'].map(colors)
```

```python
iris.head()
```

```python
iris.plot.scatter(x='sepal_width', y='sepal_length', color=iris['colors']);
```
