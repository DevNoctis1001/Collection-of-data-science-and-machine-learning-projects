##### Using pre-trained models

##### How this deep image predictor works?
- once the image is input, it needs to determine which of 1,000 classes of ImageNet the object and the image corresponds to.
- by default, each of the objects is given a probability.
- so that the deep image predictor sorts them based on probability.
- since we specified topK=5, it displays the object with top five highest probabilities.

```
from sparkdl import DeepImagePredictor
```

```
# input will be 'image' column of df 
# decodePredictions: making sense of the class from that sysnet file (the one we will be using)
# topK 5: we want to see top 5 predictions.

predictor = DeepImagePredictor(inputCol='image', outputCol='predicted_labels', modelName='ResNet50', decodePredictions=True, topK=5)
```

```
# transform the da`taframe
predictor_df = predictor.transform(df)
```

```
# get the prediction
predictor_df.select('predicted_labels').show(truncate=False)
```

We can see that the top result is "Typewrite Keyboard" with probability with 0.97282547, which is accurate. This pretained model ResNet50 is working well.
