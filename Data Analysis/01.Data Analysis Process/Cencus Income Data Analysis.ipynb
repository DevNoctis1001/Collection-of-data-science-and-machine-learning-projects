```python
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv("../00.datasets/adult_census_income_data.csv", delimiter=",", header=None, names=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "listing attribute"])
```

```python
#remove whitespace
df = df.apply(lambda x: x.astype(str).str.strip() if x.dtype == "object" else x)
```

```python
#check for any duplicates
sum(df.duplicated())
```
