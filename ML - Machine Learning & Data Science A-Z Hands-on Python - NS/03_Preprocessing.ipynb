#### Fillna Methods (Forwards Fill and Backwards Fill)
+ Foward Fill => if there is empty cell ahead, forwards the current value to next empty cell and fill it
+ Backward Fill => if there is empty cell behind, bring the current value to the behind cell and fill it.

We can see there is NaN cell in both row 14 and 23.

```python
df_new.fillna(method='ffill')
```

```python
df_new.fillna(method='bfill')
```

### Using sklearn Imputer - SimpleImputer

```python
from sklearn.impute import SimpleImputer
```

```python
missing_value_imputer = SimpleImputer(missing_values= np.nan, strategy='mean')
```

```python
missing_value_imputer.fit(df_new)
```

```python
# NOTE: there are missing values in row 14 and 23
df_new
```

```python
# used the imputer to fit and transform 
df2 = missing_value_imputer.transform(df_new)
df2
```

# Outliers Detection II

```python
df = pd.read_csv('./Data/Data_Set.csv', header=2)
```

```python
df = df.replace('!', np.NaN)
```

```python
df = df.fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
df = df.apply(pd.to_numeric)
```

```python
df.info()
```

#### Potential Outliers Detection

We can see there are potential an outlier in both `E_Plug` and `E_Heat`

```python
plt.figure(figsize=(12, 6))
df.boxplot();
```

```python
df.describe().transpose()
```

```python
Q1 = df['E_Plug'].quantile(0.25) # Q1 
Q1
```

```python
Q3 = df['E_Plug'].quantile(0.75) # Q3
Q3
```

```python
IQR = Q3 - Q1  # Inter Quantile Range
IQR
```

#### Mild Outliers

```python
mild_lower_bound = Q1 - (1.5 * IQR)
mild_upper_bound = Q3 + (1.5 * IQR)

mild_lower_bound, mild_upper_bound
```

#### Extreme Outliers

```python
extreme_lower_bound = Q1 - (3 * IQR)
extreme_upper_bound = Q3 + (3 * IQR)

extreme_lower_bound, extreme_upper_bound
```

```python
df['E_Plug'].max()
```

#### Outlier Replacement

The outlier point of 120 is even higher than Extreme Upper Bound of 75.

So, we gonna replace that point the value near to it.

```python
df['E_Plug'] = df['E_Plug'].replace(120, 42)
```

```python
df['E_Plug'].max()
```
