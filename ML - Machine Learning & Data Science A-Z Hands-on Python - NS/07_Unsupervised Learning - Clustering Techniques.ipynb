## K-Mean Evaluation Metrics

### Elbow Method

![](Images/kmean7.png)
![](Images/kmean8.png)

+ We want Inertia value as small as possible.
+ Inertia value is large at start, however it gets smaller 
after k value got larger.
+ However at some number of k value, inertia value doesn't get decrease anymore.
+ That point of k value is called Elbow Point or Optimal Point.

--------

## Model Development

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
data = load_iris()

X = data['data']
y = data['target']
```

```python
data['target_names']
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3) #as there are three types of flowers in this scenarios
```

```python
kmeans.fit(X)
```

```python
predictions = kmeans.predict(X)
```

```python
predictions
```

## Find Cluster Center point

```python
centeriods = kmeans.cluster_centers_
centeriods
```

## Visualize Predictions

```python
data['feature_names']
```

```python
# Petal Length and Petal Width based original labels
plt.scatter(X[:, 2], X[:, 3], c=y);
plt.xlabel('Petal Length(cm)')
plt.ylabel('Petal Width(cm)');
plt.title('Petal Length and Petal Width based original labels');
```

```python
# Petal Length and Petal Width based on predicted labels
plt.scatter(X[:, 2], X[:, 3], c=predictions);
plt.scatter(centeriods[:, 2], centeriods[:, 3], marker='o', color='red', s=120); # centeriods
plt.xlabel('Petal Length(cm)')
plt.ylabel('Petal Width(cm)');
plt.title('Petal Length and Petal Width based on predicted labels');
```

## Model Evaluation

```python
# get the model's inertia
inertia = kmeans.inertia_
inertia
```

```python
# find the best optimal value based on elbow method
k_inertia = [] # intialize the place holder
n_cluster = 10
for i in range(1, n_cluster):
    model = KMeans(n_clusters=i, random_state=44)
    model.fit(X)
    k_inertia.append(model.inertia_)
```

```python
plt.plot(range(1, n_cluster), k_inertia, color='green', marker='o');
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method - Inertia visualization');
```

We can see that around k=5 or 6, there is no significant reduction of Inertia value.

# DBSCAN - Density Based Technique

![](Images/dbscan1.png)
![](Images/dbscan2.png)

## Model Development

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data['data']
y = data['target']
```

```python
from sklearn.cluster import DBSCAN
```

```python
db_model = DBSCAN(eps=0.5, min_samples=4)
```

```python
db_model.fit(X)
```

## Get the labels from DBSCAN model

```python
predictions = db_model.labels_
predictions
```

## Visualize Predictions

```python
# Petal Length and Petal Width based original labels
plt.scatter(X[:, 2], X[:, 3], c=y);
plt.xlabel('Petal Length(cm)')
plt.ylabel('Petal Width(cm)');
plt.title('Petal Length and Petal Width based original labels');
```

```python
# Petal Length and Petal Width based on predicted labels
plt.scatter(X[:, 2], X[:, 3], c=predictions);
plt.xlabel('Petal Length(cm)')
plt.ylabel('Petal Width(cm)');
plt.title('Petal Length and Petal Width based on predicted labels');
```

# Linkage - Hierarchical Based Technique

![](Images/hierar1.png)

## Three types of Linkage Methods 
![](Images/hierar2.png)

## Model Development

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data['data']
y = data['target']
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
hierarchy_model = linkage(X, method='complete')
```

```python
dnd = dendrogram(hierarchy_model)
```

```python
predictions = fcluster(hierarchy_model, t=4, criterion='distance')
```

## Visualize Predictions

```python
# Petal Length and Petal Width based on predicted labels
plt.scatter(X[:, 2], X[:, 3], c=predictions);
plt.xlabel('Petal Length(cm)')
plt.ylabel('Petal Width(cm)');
plt.title('Petal Length and Petal Width based on predicted labels');
```
