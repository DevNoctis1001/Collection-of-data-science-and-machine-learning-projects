# 1) Dropping Columns and Rows

## Dropping rows
- df.drop(rownumber, axis = 0)
- df.drop(labels = [row_numbers_list])

```python
data.drop(0)
```

```python
data.drop(labels = [0,2,4])
```

## Drop column
- df.drop(col_name, axis = 1)
- df.drop(columns = [col_name_list]]

```python
data.drop('id', axis = 1)
```

```python
# if we don't want to remember the axis, then use the following alternative approach
data.drop(columns = ["id"])
```

```python
data.drop(columns = ["id", "artistId", "title"])
```

---------

dropping won't change the original data frame unless you tell them. 

```python
data.head()
```

## to make changes on original data frame, use `inplace` keyword

```python
data.drop(columns = ["id"], inplace = True)
data.head()
```

Dropping columns are useful when you have existing dataframe you have already imported.
but you can also import only necessary columns during importing in the first place.

--------------

## selection of only specific columns during importing from csv, `usecols`
pd.read_csv("test.csv, usecols = [list_of_columns])

```python
data = pd.read_csv("Data/artwork_sample.csv", usecols = ["artist", "title"])
```

# 2) Changing Columns Casing

```python
data = pd.read_csv("Data/artwork_sample.csv")
```

```python
data.columns
```

## change the casing of column name
- data.columns.str.lower()
- data.columns = [x.lower() for x in data.columns]
- data.columns = map(lambda x : x.lower(), data.columns)
- using regex 

```python
# directly using lower method
data.columns.str.lower()
```

```python
# alternative approach: loop through 
[x.lower() for x in data.columns]
```

```python
data.columns
```

```python
# to make changes, need to assign
data.columns = [x.lower() for x in data.columns]
```

```python
data.columns
```

### 3rd approach using map

```python
data = pd.read_csv("Data/artwork_sample.csv")
```

```python
data.columns
```

```python
# using map
data.columns = map(lambda x : x.lower(), data.columns)
```

```python
data.columns
```

### Changing specific columns using `regex`

```python
data = pd.read_csv("Data/artwork_sample.csv")
```

```python
data.columns
```

```python
import re
```

```python
data.columns = [re.sub(r'([A-Z])', r'_\1', x).lower() for x in data.columns]
```

```python
data.columns
```

# 3) Renaming Columns
- df.rename(columns = {old_col: new_col})
- df.rename({old_col : new_col}, axis = 1)
- df.rename(columns = lambda x: x.lower(), inplace = True)
- df.columns = [column_name_list]
- pd.read_csv("test.csv", header = 0, names = [list_of_columns_name_that_we_want_to_use])

```python
data = pd.read_csv("Data/artwork_sample.csv")
```

```python
data.columns
```

```python
data.rename(columns = {"thumbnailUrl": "thumbnail"})
```

```python
# alternative approach
data.rename({"thumbnailUrl": "thumbnail"}, axis = 1)
```

```python
# using function ()
data.rename(columns = lambda x: x.lower(), inplace = True)
```

```python
data.columns
```
