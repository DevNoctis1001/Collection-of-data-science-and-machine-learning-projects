# 2) Replacing Bad data with NaN

```python
data = pd.read_csv("Data/artwork_data.csv", low_memory = False)
```

```python
data.head(2)
```

## 2.1) replacing with NaN  using `replace` method, for `dateText` where `date not known`

### first check the the entire column for na values

```python
pd.isna(data.loc[:, "dateText"])
```

### replacing with nan

```python
from numpy import nan
```

```python
data.replace({'dataText' : {'date not known' : nan}}, inplace = True)
```

```python
data.head(2)
```

## 2.2) replacing nan using `loc` method

```python
data = pd.read_csv("Data/artwork_data.csv", low_memory = False)
data.head(2)
```

```python
data.loc[data.dateText == "date not known", ["dateText"]] = nan
```

```python
data.head(2)
```

----------

### check `year` data for bad data

```python
# check year data which contains values but contains data other than digits too
data.loc[data.year.notnull() & data.year.str.contains("[^0-9]")]
```

```python
data.loc[data.year.notnull() & data.year.astype(str).str.contains("[^0-9]"), ["year"]] = nan
```

```python
data.head(2)
```

# 4) Dropping rows of data

```python
data = pd.read_csv("Data/artwork_data.csv", low_memory = False)
data.head(2)
```

```python
data.shape
```

### 4.1) drop nan values

```python
data.dropna()
```

```python
data.dropna().shape
```

### 4.2) dropping rows only if all values of all columns are null
- useful at the final stage of data cleaning, after you converted all values to numerical

```python
data.dropna(how = "all").shape
```

### 4.3) dropping rows using `threshold`

```python
# drop the row only if 15 of columns values are null (out of data frame's 20 columns)
data.dropna(thresh = 15). shape
```

## 4.4) dropping rows based on specific column's values are NaN
- or
- and (using `how = all`)

```python
#drop the rows only if year or acquisitionYear column values are NaN
data.dropna(subset = ["year", "acquisitionYear"]).shape
```

```python
#drop the rows only if BOTH year and acquisitionYear column values are NaN
data.dropna(subset = ["year", "acquisitionYear"], how = "all").shape
```

```python
data.dropna(subset = ["year", "acquisitionYear"], how = "all", inplace = True)
```

# 5) Identifying and Dropping Duplicate Data

```python
data = pd.read_csv("Data/artwork_sample.csv")
```

```python
data.head(2)
```

### 5.1) dropping duplicates based on specific column

```python
# now we try to drop duplicates based on "artist" column
data.drop_duplicates(subset = ["artist"])
```

now only two rows remains. This makes sense because there are only 2 distinct artist name.

### 5.2) keep first and keep last

```python
data.drop_duplicates(subset = ["artist"], keep = "first")
```

```python
data.drop_duplicates(subset = ["artist"], keep = "last")
```

### 5.3) keep none if there is more than one duplicate for that value

```python
data.drop_duplicates(subset = ["artist"], keep = False)
# this returns none because each of the artist has more than one rows
```

-------
