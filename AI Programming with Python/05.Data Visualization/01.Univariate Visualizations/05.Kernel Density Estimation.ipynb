```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = [0.0, 0.3, 4.5, 8.0]
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.kdeplot(x=data);

plt.subplot(1,2,2)
sns.rugplot(x=data, color='r');
```

```python
plt.figure(figsize=(12, 5));

sns.displot(data=data, kind='kde', rug=True, rug_kws={'color':'r'});
```

```python
sns.displot(data=data, kind='kde', rug=True, rug_kws={'color':'r'}, bw_adjust=0.1);
```
