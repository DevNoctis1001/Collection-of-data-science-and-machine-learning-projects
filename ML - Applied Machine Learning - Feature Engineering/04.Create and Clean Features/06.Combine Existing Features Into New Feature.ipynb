## Creating And Cleaning Features: Combine Existing Features Into New Feature

### Read In Data

```python
# Read in data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

titanic_df = pd.read_csv('../Data/titanic_cabin_ind.csv')
titanic_df.head()
```

### Combine `SibSp` & `Parch` Into New `Family_cnt` Feature

```python
# Generate categorical plots for ordinal features
for col in ['SibSp', 'Parch']:
    sns.catplot(x=col, y='Survived', data=titanic_df, kind='point', aspect=2)
    plt.ylim(0, 1)
```

```python
# Create a new family count feature
titanic_df['Family_cnt'] = titanic_df['SibSp'] + titanic_df['Parch']
sns.catplot(x='Family_cnt', y='Survived', data=titanic_df, kind='point', aspect=2, )
plt.ylim(0, 1)
```

```python
# Create new CSV with updated data
titanic_df.to_csv('../Data/titanic_family_cnt.csv', index=False)
```
