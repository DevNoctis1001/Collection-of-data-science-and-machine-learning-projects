## Prepare Features For Modeling: Standardize All Features

### Read In Data

```python
# Read in data
import pandas as pd
from sklearn.preprocessing import StandardScaler

titanic_train = pd.read_csv('../Data/Split_Data/train_features.csv')
titanic_val = pd.read_csv('../Data/Split_Data/val_features.csv')
titanic_test = pd.read_csv('../Data/Split_Data/test_features.csv')

titanic_train.head()
```

-----------

### Standardize Features

What happens when you are fitting the scaler?
- it's `computing the mean and standard devision` for `each individual features` in our training data.

```python
# Fit the scaler on the training data
scaler = StandardScaler()
scaler.fit(titanic_train)
```

For tranformation,
- we need tell the scaler explicit columns we want to transform.
- based on the fit data (trained data), by using that to transform each value for that feature in the training, validation and test sets.

```python
# Scale the training, test, and validation sets

# get all features columns name
features = titanic_train.columns

# transformation
titanic_train[features] = scaler.transform(titanic_train[features])
titanic_val[features] = scaler.transform(titanic_val[features])
titanic_test[features] = scaler.transform(titanic_test[features])

titanic_train.head()
```

 now we can see the above values are on the same scale using standard deivsions.
